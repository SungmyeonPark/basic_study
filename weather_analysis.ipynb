```python
import requests
```

```python
from bs4 import BeautifulSoup as BS
```

HTTP 요청을 보내는 Request 라이브러리

Beautiful Soup은 HTML 코드를 XML형식으로 변환하여 웹을 탐색할때 유용하게 사용되는 라이브러리

```python
url = 'http://www.weather.go.kr/weather/observation/currentweather.jsp'
```

```python
response = requests.get(url)
```

기상청의 관측자료의 주소로 정보를 요청해 response에 저장

```python
if response.status_code != 200:
    print("%d 에러가 발생했습니다." % response.status_code)
    quit()
```

```python
print(response.content)
```

받아온 정보를 그대로 보여주면 위와 같음

```python
soup = BS(response.content, 'html.parser')
```

```python
print(soup)
```

Beutifulsoup을 이용해서 정리한 화면의 HTML코드

```python
table = soup.find('table', {'class' : 'table_develop3'})
```

```python
print(table)
```

```python
data = []
for tr in table.find_all('tr'):
    tds = list(tr.find_all('td'))
    for td in tds:
        if td.find('a'):
            region = td.find('a').text
            temperature = tds[5].text
            humidity = tds[9].text
            data.append([region, temperature, humidity])
```

TR은 Table Row의 약자로 HTML에서 행을 그릴때 쓰는 코드, TD는 Table data의 약자로 셀(열)을 그릴때 쓰는 코드
테이블에서 TR을 하나씩 꺼내서 TR변수에 저장하고 각 셀 값만 찾아서 TDS에 배열로 저장후 TD를 하나씩 꺼내서 변수에 저장하고 a(지역 이름)을 찾아서 region에 저장 6번째 기온과 10번째 습도를 찾아서 각각 변수에 저장훈 data라는 배열 변수에 저장 

```python
print(data)
```

```python
with open('weather.csv', 'w') as file:
    file.write('지역, 기온, 습도\n')
    for i in data:
        file.write('{0}, {1}, {2}\n'.format(i[0], i[1], i[2]))
```
